```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_parquet('../data/features_x1.parquet')
data.head()
```

Gerando histogramas para as features (exceto histogramas H1 a H45)

```python
columns = data.columns
colors = sns.color_palette("hls", 11)

for aux, index in enumerate(columns[:11]):
    fig = plt.figure(figsize=(10, 8))
    sns.histplot(data=data, x=index, color=colors[aux], kde=True, linewidth=0.1)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.ylabel("Ocorrências")
    plt.legend({"Estimava de de Densidade por Kernel (KDE)"})
    plt.savefig(f'img/{index}_histplot.png', format='png')
    plt.close(fig)
```
